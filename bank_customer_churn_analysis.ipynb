```python
# gini
from sklearn.metrics import roc_curve, auc 
def proc_gini(y,z):
    fpr, tpr, _ = roc_curve(y, z, pos_label=0)
    roc_gini = (auc(fpr, tpr)-0.5)*2
    return roc_gini 
```

## single variable bin to find cutoffs

```python
def cal_split_to_bin(df,col,splits):
    col_bin = col+'_bin'
    df[col_bin] = pd.cut(df[col], splits, include_lowest=True, duplicates = 'drop')  # 切割数据，离散化连续数据
    if df[col].count()<df.shape[0]:
        df[col_bin] = df[col_bin].cat.add_categories(['missing'])
        df[col_bin] = df[col_bin].fillna('missing')
    return df
```

```python
col_var = 'age'
split_bin = [0,20,25,30,35,40,48,52,56,58,60,65,999]
df1 = cal_split_to_bin(df_train, col_var,split_bin)
```

```python
col = col_var+'_bin'
label_d = 'exited'  # target

total_bad_num = df_train[label_d].sum()
total_good_num = len(df_train) - df_train[label_d].sum()
    
sample1 = df1[[col]].sort_values(by=col).drop_duplicates().reset_index(drop=True).reset_index().rename(columns={'index':'bucket'})

badnum_df = df1.groupby([col])[label_d].sum().reset_index().rename(columns = {label_d:'bad_num'})

num_df = df1.groupby([col])['age'].count().reset_index().rename(columns = {'age':'total'})

sample1['name'] = col_var
# sample.columns = ['bucket','bin','name']

sample1 = sample1.merge(badnum_df[[col,'bad_num']],on=col,how='left')
sample1 = sample1.merge(num_df[[col,'total']],on=col,how='left')

sample1['badnum_rate'] = sample1['bad_num']/sample1['total']
sample1['cum_badnum'] = sample1['bad_num'].cumsum()

sample1['cum_total'] = sample1['total'].cumsum()
sample1['cum_badnum_rate'] = sample1['cum_badnum']/sample1['cum_total']
sample1['gain'] = (sample1['bad_num']/total_bad_num).cumsum()
sample1
```

```python
sample1.plot('age_bin', 'cum_badnum_rate')
plt.xticks(rotation = 30) # adjust font rotation to show all bins clearly
plt.title('Accumulated Churn Rate by Age')
plt.show()
```

<span style="color:red;font-family: Arial;">It shows people betweem 40 and 52 years old, they have much higher exited probablity than others. Bank can do some activities for this segment to decrease churn rate</span>
